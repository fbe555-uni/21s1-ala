### Problem 1
Given the subspace  
{{r'$\left\{' f"{LVector(['s-2t', 's+t', '3t'])}" r': \text {s, t in }\Bbb R \right\}$'}}  

#### a) find a basis

To find a basis for this subspace, the defining vector can be rewritten as two vectors with the two variables as factors:  
{{'$' f" {LVector(['s-2t', 's+t', '3t'])} = {LVector(['s', 's', '0'])} + {LVector(['-2t', 't', '3t'])} = s {LVector([1, 1, 0])} + t {LVector([-2, 1, 3])} $" }}  
Now, since s and t are any numbers in $\Bbb R$, this is the same as  
{{r'$Span\left\{' f"{LVector([1, 1, 0])}, {LVector([-2, 1, 3])}" r'\right\}$'}}  
Now since these two vectors are not multiples of each other, they form a basis

#### b) state the dimension.  

Since the basis has two elements, the subspace has dimension 2

### Problem 13
Determine the dimensions of $\text{Nul }A$ and $\text{Col }A$ for  
{{LMatrix([[1,-6,9,0,-2],[0,1,2,-4,5],[0,0,0,5,1],[0,0,0,0,0]], name='A', wrapper='$$')}}

The dimensions of both of these subspaces can be determined from the number of pivots in A. The given matrix is already in echelon form, and has three pivots.  
Now from Theorem 4.14 The Rank Theorem, it is given that:  
$\text{dim Nul } A = \#non-pivotcolumns = 2$  
and  
$\text{dim Col } A = \#pivotcolumns = 3$  


### Problem 19
$V$ is a vector space. Mark each statement True or False. Justify each answer.

#### a) The  number  of  pivot  columns  of  a  matrix  equals  the dimension of its column space.

__True__: This is stated in theorem 4.14 The rank theorem

#### b) A plane in $\Bbb R^3$ is a two-dimensional subspace of $\Bbb R^3$.

__False__, The plane has to include the origin for it to be a subspace (must contain the zero vector)

#### c) The dimension of the vector space $\Bbb P^4$ is 4.

__False__, The dimension of $\Bbb P^n$ is $n+1$ since the polynomial also has a constant element

#### d) If $\text{dim }V = n$ and $S$ is a linearly independent set in $V$, then $S$ is a basis for $V$.

__False__, This is only the case if $S$ has $n$ elements

#### e) If  a  set $\{v_1, \dots, v_p\}$ spans  a  finite-dimensional  vector space $V$ and if $T$ is a set of more than $p$ vectors in $V$, then $T$ is linearly dependent.

__True__, A set of vectors must be linearly dependent if it contains more vectors than the dimension of the vector space they belong to. It is not specified, that the given vector set is a basis for $V$, only that it spans it. This means that however $V$ does not necessarily have dimension $p$ since we don't know that the vector set spanning $V$ is linearly independent (it is not necessarily a basis for $V$).
